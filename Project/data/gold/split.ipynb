```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("gold_Qwen3.7_max.csv")

```

```python
df.dropna(subset=['sarcasm_label'], inplace=True)
```

```python
df.info()
```

```python
stratify_keys = (
	df['sarcasm_label'].fillna('NA').astype(str)
	+ '___'
	+ df['sarcasm_type'].fillna('NA').astype(str)
)

train_df, test_df = train_test_split(
	df, test_size=0.1, stratify=stratify_keys, random_state=42
)

# 3. Lưu thành 2 file riêng biệt
train_df.to_csv("train_data.csv", index=False)
test_df.to_csv("test_data.csv", index=False)

print(f"Số mẫu Train: {len(train_df)}")
print(f"Số mẫu Test: {len(test_df)}")
```
