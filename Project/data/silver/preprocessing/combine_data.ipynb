```python
df =  df[['platform', 'video_id', 'video_url', 'video_title',
          'video_core_content','video_target_sentiment','comment']]
df_tt = df_tt[['Video ID', 'Nội dung chính toàn video',
       'Các điểm nhấn/Tranh cãi (Key Highlights)']]
df_yt = df_yt[['Video ID', 'Nội dung chính toàn video',
       'Các điểm nhấn/Tranh cãi (Key Highlights)']]
```

```python
df_tt['Video ID'] = df_tt['Video ID'].astype(str)
df_yt['Video ID'] = df_yt['Video ID'].astype(str)
df['video_id'] = df['video_id'].astype(str)
```

```python
df_combined = pd.concat([df_tt, df_yt], ignore_index=True)

Core_content = df_combined.set_index('Video ID')['Nội dung chính toàn video']
highlight = df_combined.set_index('Video ID')['Các điểm nhấn/Tranh cãi (Key Highlights)']
df['video_core_content'] = df['video_id'].map(Core_content)
df['video_highlight'] = df['video_id'].map(highlight)
```
