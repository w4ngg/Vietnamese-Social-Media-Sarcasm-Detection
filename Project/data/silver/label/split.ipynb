```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
DATA_PATH = '/home/check/DATA/university/yr3, hk2/IE403 - Khai thác dữ liệu truyền thông xã hội/Data/silver/silver1.csv'

# Tạo thư mục lưu sample
output_dir = 'labeling_samples'
os.makedirs(output_dir, exist_ok=True)
```

```python
print("📊 Đang load dữ liệu...")
df = pd.read_csv(DATA_PATH)
print(f"✅ Load thành công! Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
```

```python
np.random.seed(42)  # Để reproducible

# 1. Tạo 3 bản sample 50 mẫu GIỐNG NHAU (dùng để đo độ đồng thuận IAA)
print("\n🔄 Tạo 3 bản sample 50 mẫu giống nhau...")

sample_50 = df.sample(n=50, random_state=42).copy()

for i in range(1, 4):
    filename = f"{output_dir}/sample_50_same_set_{i}.csv"
    sample_50.to_csv(filename, index=False)
    print(f"✅ Đã tạo sample 50 mẫu (bản {i}) → {filename}")

# 2. Tạo 3 bản sample 100 mẫu KHÁC NHAU (dùng để label chính)
print("\n🔄 Tạo 3 bản sample 100 mẫu khác nhau...")

for i in range(1, 4):
    # Dùng random_state khác nhau để có 3 tập khác nhau
    sample_100 = df.sample(n=100, random_state=100 + i*10).copy()
    filename = f"{output_dir}/sample_100_set_{i}.csv"
    sample_100.to_csv(filename, index=False)
    print(f"✅ Đã tạo sample 100 mẫu (bản {i}) → {filename}")
```

```python
print("\n" + "="*60)
print("📈 THỐNG KÊ SAMPLING")
print("="*60)
print(f"Tổng số mẫu gốc          : {len(df):,}")
print(f"Đã tạo 3 bản 50 mẫu giống nhau (IAA test)")
print(f"Đã tạo 3 bản 100 mẫu khác nhau (Label chính)")
print(f"Files được lưu tại thư mục: ./{output_dir}/")

# Xem trước một sample
print("\n🔍 Preview 5 dòng của sample 50 mẫu:")
print(sample_50[['platform', 'comment', 'video_target_sentiment']].head())

print("\n🎉 Hoàn thành! Bạn có thể bắt đầu label.")
```
